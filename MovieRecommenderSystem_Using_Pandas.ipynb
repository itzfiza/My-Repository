**Applying functions**

```python
def rating_fun(x):

  if(x > 8.0):
    return "Must watch"
  elif(x > 6.0 and x < 8.0):
    return "Average"
  else:
    return "Not recommended"
```

```python
movies_df["user_review"] = movies_df['rating'].apply(rating_fun)
movies_df.head()
```
